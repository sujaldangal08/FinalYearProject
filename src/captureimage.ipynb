```python
import cv2

# Initialize the webcam (usually, 0 is the default camera)
camera = cv2.VideoCapture(0)

# Create a face cascade classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Set the number of face images you want to capture
num_face_images_to_capture = 5  # Change this to the desired number

# Create a variable to keep track of the captured face image count
captured_face_images = 0

while captured_face_images < num_face_images_to_capture:
    # Read a frame from the camera
    ret, frame = camera.read()

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Iterate over detected faces and capture them
    for (x, y, w, h) in faces:
        # Crop the detected face
        face = frame[y:y+h, x:x+w]

        # Display the detected face
        cv2.imshow('Detected Face', face)

        # Capture the face image when the "c" key is pressed
        key = cv2.waitKey(1)
        if key == ord('c'):
            # Save the captured face image to a file with a unique name
            image_filename = f'captured_face_{captured_face_images + 1}.jpg'
            cv2.imwrite(image_filename, face)
            print(f'Face Image {captured_face_images + 1} saved as {image_filename}')
            captured_face_images += 1

    # Exit the loop when the "q" key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close the OpenCV window
camera.release()
cv2.destroyAllWindows()
```
